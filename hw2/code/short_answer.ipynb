# Short answers

```python
import numpy as np
import cv2
from scipy.signal import convolve2d
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("../data/cat_bread.jpg") 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = gray.astype(np.float64)
```

```python
# spatial gradient (x direction)
Kx = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

grad_x = convolve2d(gray, Kx, mode='same', boundary='symm')
```

```python
# spatial gradient (y direction)
Ky = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
])

grad_y = convolve2d(gray, Ky, mode='same', boundary='symm')
```

```python
# averaging filter

Kavg = np.ones((5,5)) / 25

avg_img = convolve2d(gray, Kavg, mode='same', boundary='symm')
```

```python
# visualization
from pathlib import Path
import matplotlib.pyplot as plt

OUT_DIR = Path("../output")
OUT_DIR.mkdir(exist_ok=True)

plt.imshow(gray, cmap='gray')
plt.title("Original Grayscale")
plt.savefig(OUT_DIR / "originial.png", dpi=300, bbox_inches="tight")
plt.show()

plt.imshow(grad_x, cmap='gray')
plt.title("Gradient X (Sobel)")
plt.savefig(OUT_DIR / "grad_x.png", dpi=300, bbox_inches="tight")
plt.show()

plt.imshow(grad_y, cmap='gray')
plt.title("Gradient Y (Sobel)")
plt.savefig(OUT_DIR / "grad_y.png", dpi=300, bbox_inches="tight")
plt.show()

plt.imshow(avg_img, cmap='gray')
plt.title("5x5 Average Filter")
plt.savefig(OUT_DIR / "average.png", dpi=300, bbox_inches="tight")
plt.show()
```
